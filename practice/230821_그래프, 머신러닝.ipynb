```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
iris
```

```python
data = iris['data']
```

```python
target = iris['target']
```

```python
target
```

```python
iris['feature_names']
```

```python
iris['target_names']
```

```python
import pandas as pd

iris_df = pd.DataFrame(data, columns = iris['feature_names'])
```

```python
iris_df['label'] = iris.target
```

```python
iris_df.head()
```

```python
# train : 머신러닝이 학습하는 데이터
# test  : 머신러닝 학습이 끝나고 잘 했는지 평가하는 데이터
X_train, X_test, y_train, y_test = train_test_split(iris['data'],
                                                    iris['target'],
                                                    test_size = 0.2,
                                                    random_state = 11) # 난수 고정
```

```python
train, test = train_test_split(iris_df,
                              test_size = 0.2,
                              random_state = 11)
```

```python
# 분류 알고리즘 호출
dt_clf = DecisionTreeClassifier(random_state=11)

# 학습 - 학습 데이터
dt_clf.fit(X_train, y_train)
```

```python
# 예측(분류) - 테스트 데이터
pred = dt_clf.predict(X_test)
pred # 예측값
```

```python
# 평가 - 머신러닝이 잘 학습했는지를 수치로 확인하겠다
from sklearn.metrics import accuracy_score
# pred : 예측값, y_test : 실제값
accuracy_score(y_test, pred)
```

## 머신러닝 단계
* 1. 머신러닝에 학습할 데이터 준비
  - 1.1. 데이터를 가져오는 단계
  - 1.2. 학습할 수 있는 모양으로 바꾸는 단계
* 2. 알고리즘
  - 머신러닝 학습 취지에 맞는 알고리즘 선택(분류, 회귀, 군집)
  - 최적의 알고리즘 학습 방법을 찾기
* 3. 학습
  - 학습 데이터를 사용해서 진행
* 4. 예측
  - 테스트 데이터를 사용해서 예측 결과 값 추출
* 5. 평가
  - 예측 결과 값을 토대로 머신러닝 평가
  - 평가 기준이 낮으면 다른 학습 방법을 시도(2번으로 돌아가기)

```python
from sklearn.model_selection import cross_val_score

iris_data = load_iris()
dt_clf = DecisionTreeClassifier(random_state= 156)
data = iris_data.data
label = iris_data.target

# 원래 방법
dt_clf.fit(data, label)

# cross_val_score
score = cross_val_score(dt_clf, data, label, scoring='accuracy', cv = 5)
```
