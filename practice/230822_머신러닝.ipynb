```python
# LabelEncoder : 한 컬럼 문자 > 한 컬럼의 숫자, 보통 종속변수(Label), 의외로 이지선다 (남/녀)를 나눌때는 얘를 쓴다.
# OneHotEncoder : 한 컬럼 문자 > 여러 컬럼의 숫자, 보통 독립변수(X, Feature), 거의 얘를 써서 바꾼다.
# 성별 : LabelEncoder, 선착장 : OneHotEncoder 쓰겠다.

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

lb_enc = LabelEncoder()
oh_enc = OneHotEncoder()

sex_enc = lb_enc.fit_transform(titanic_df['Sex'])
emb_enc = oh_enc.fit_transform(titanic_df[['Embarked']])
```

```python
def get_category(age):
  cat = ''
  if age <= -1:
    cat = 'Unknown'
  elif age <= 5:
    cat = 'Baby'
  elif age <= 12:
    cat = 'Child'
  elif age <= 18:
    cat = 'Teenager'
  elif age <= 25:
    cat = 'Student'
  elif age <= 35:
    cat = 'Young Adult'
  elif age <= 60:
    cat = 'Adult'
  else :
    cat = 'Elderly'
  return cat
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()

dt_clf.fit(X_train.values, Y_train)

pred = dt_clf.predict(X_test.values)
```

```python
pred
```

```python
# 정확도
from sklearn.metrics import accuracy_score

print('정확도 :', accuracy_score(Y_test, pred))
```

```python
# 오차행렬
from sklearn.metrics import confusion_matrix

print('오차행렬\n', confusion_matrix(Y_test, pred))
```

```python
# 정밀도 : 0 ~ 1 값을 가지고 1에 가까울수록 정밀도가 높다
from sklearn.metrics import precision_score
# 재현율 : 0 ~ 1 값을 가지고 1에 가까울수록 재현율이 높다
from sklearn.metrics import recall_score

print('정밀도 :', precision_score(Y_test, pred))
print('재현율 :', recall_score(Y_test, pred))
```

```python
# f1_score : 정밀도와 재현율 모두 높을때 좋은 점수가 나온다, 0 ~ 1 사이의 값
from sklearn.metrics import f1_score

print('F1_Score :', f1_score(Y_test, pred))
```

```python
# roc_auc_score
from sklearn.metrics import roc_auc_score

print('ROC_AUC :',roc_auc_score(Y_test, pred))
```
